# Scaling

```python
data_minmax = data1
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
scale_columns = ["overall","Sentiment","h"]
data_minmax[scale_columns] = min_max_scaler.fit_transform(data1[scale_columns])
data_minmax.head()
```

# RAW data

```python
import matplotlib.pyplot as plt
data_minmax.plot.scatter(x= 'overall',y = 'h')
plt.title(" Raw Data")
plt.show()
```

# K-Mean clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python
distorsions = []
df = data_minmax[scale_columns]
for k in range(2, 10):
  kmeans = KMeans(n_clusters=k)
  kmeans.fit(df)
  distorsions.append(kmeans.inertia_)
fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 10), distorsions)
plt.grid(True)
plt.title('Elbow curve');
```

Optimum value of k is identified as 3

# Silhouette score:

```python
#Note don't ever try to run the code:Takes long time
import matplotlib.cm as cm
range_n_clusters = [3,4]
for n_clusters in range_n_clusters:
  clusterer = KMeans(n_clusters=n_clusters, random_state=10)
  cluster_labels = clusterer.fit_predict(df)
  silhouette_avg = silhouette_score(df, cluster_labels)
  print("For n_clusters =", n_clusters, "The average silhouette_score is :", silhouette_avg)
```

# Cluster number 4

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
kmeans = KMeans(n_clusters=4,n_init=4).fit(df)
#n_clusters=K;n_init refers to the number of times k=mean algo will be running wth diff centroid
centroids = kmeans.cluster_centers_#########################Centroid axis
labels=kmeans.labels_
print(centroids)
print(labels)
print('Attributes used for cluster model: ', scale_columns)
```

```python
from mpl_toolkits.mplot3d import Axes3D 
fig = plt.figure(figsize=(5,5))
ax = Axes3D(fig)
x=df['overall']
y=df["h"]
z=df["Sentiment"]
ax.scatter(x, y, z, c=kmeans.labels_.astype(float), marker='o',cmap='Paired',s=10, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
ax.set_xlabel('Overall')
ax.set_ylabel('Helpfulness')
ax.set_zlabel('Sentiment')
plt.title("Overall VS Helpfulness VS Bonus Sentiments")
plt.show()
```
