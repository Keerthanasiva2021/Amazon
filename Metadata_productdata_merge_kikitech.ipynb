```python
#Libraries
import pandas as pd
import gzip
import json
```

```python
#importing json file as df
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF("C:\\Users\\keesi\\Downloads\\meta_All_Beauty (2).json.gz")
```
